```python
train, test = train_test_split(df_main, test_size=0.20, random_state=41)
```

```python
print("Size of train data: ", train.shape)
print("Size of test data: ", test.shape)
```

```python
pv_ = pd.pivot_table(train, index='reviews_username', columns='id', values='reviews_rating', aggfunc='mean').fillna(0)
pv_
```

```python
# Copy the train dataset into dummy_train
dummy_train = train.copy()
```

```python
# The movies not rated by user is marked as 1 for prediction. 
dummy_train['reviews_rating'] = dummy_train['reviews_rating'].apply(lambda x: 0 if x>=1 else 1)
```

```python
# Convert the dummy train dataset into matrix format.
dummy_train = dummy_train.pivot_table(
    index='reviews_username',
    columns='id',
    values='reviews_rating'
).fillna(1)
```

```python
from sklearn.metrics.pairwise import pairwise_distances

# Creating the User Similarity Matrix using pairwise_distance function.
user_correlation = 1 - pairwise_distances(pv_, metric='cosine')
user_correlation[np.isnan(user_correlation)] = 0
print(user_correlation)
```
